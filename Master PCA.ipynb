```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
data = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(data)
```

```python
scaled_data = scaler.transform(data)
```

```python
data = pd.DataFrame(scaled_data,columns = cancer['feature_names'])
data.head(1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

```python
pca.fit(data)
```

```python
x_pca = pca.transform(data)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x_pca[:,0],x_pca[:,1], c = cancer['target'])
plt.xlabel('First component')
plt.ylabel('Second component')
plt.title('PCA Scatter plot', size = 15);
```

```python
weight_analysis = pd.DataFrame(pca.components_,columns = cancer['feature_names'])
weight_analysis
```

```python
plt.figure(figsize = (15,6))
sns.heatmap(weight_analysis).set(title = 'Weight of each individual variable in the PCAs extracted');
```
